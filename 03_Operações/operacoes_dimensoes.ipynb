Importando as Bibliotecas e Criando o DataSet

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({ 'ID': ['XX1', 'XX1', 'XX2', 'XX2', 'XX3', 'XX4', 'XX4', 'XX4'],
                    'DT_CONSUMO': ['2020-11-03', '2020-11-15', '2020-11-18', '2020-12-16', '2020-11-06', '2020-12-18', '2020-11-18', '2020-11-04'], 
                    'VL_CONSUMO': [50, 21, 33, 23, 45, 46, 12, 34],
                    'ID_VENDEDOR': ['X123', 'X234', 'X123', 'X123', 'X234', 'X536', 'X234', 'X123'],
                    'NOTA_NPS': [9, 6, 8, 7 ,7 ,10, 6, 8]})
df
```

```python
df_select = df[df['ID_VENDEDOR'].map(df['ID_VENDEDOR'].value_counts() > 1)]
df_select
```

```python
df_groupby = df_select[['ID_VENDEDOR', 'NOTA_NPS']].groupby(by='ID_VENDEDOR').agg({
    'NOTA_NPS': 'mean',
    })
df_groupby
```

```python
df_final = pd.merge(df, df_groupby, left_on='ID_VENDEDOR', right_on='ID_VENDEDOR', how='left').rename(columns={'NOTA_NPS_x': 'NOTA_NPS', 'NOTA_NPS_y': 'MÉDIA_NPS_VENDEDOR'})
df_final
```

```python
df = pd.DataFrame({ 'ID': ['XX1', 'XX1', 'XX2', 'XX2', 'XX3', 'XX4', 'XX4', 'XX4'],
                    'DT_CONSUMO': ['2020-11-03', '2020-11-15', '2020-11-18', '2020-12-16', '2020-11-06', '2020-12-18', '2020-11-18', '2020-11-04'], 
                    'VL_CONSUMO': [50, 21, 33, 23, 45, 46, 12, 34],
                    'CEP_MORADIA': ['22430095', '22430095', '22432240', '22432240', '21678190', '21654570', '21654570', '21654570'],})
df
```

```python
df['CEP_3DIG'] = df['CEP_MORADIA'].map(lambda x: int(str(x)[:3]))
df
```

```python
df_groupby = df[['VL_CONSUMO', 'CEP_3DIG']].groupby(by='CEP_3DIG').agg({
    'VL_CONSUMO': 'mean',
    })
df_groupby
```

```python
df_final = pd.merge(df, df_groupby, left_on='CEP_3DIG', right_on='CEP_3DIG', how='left')
df_final
```
